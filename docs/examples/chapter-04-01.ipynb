## 4.3 : First-Order Systems

### Skill-Assessment Exercise 4.2

A system has a transfer function:
$$G\left( s \right) = \frac{{50}}{{s + 50}}$$

Find the time constant, settling time, and rise time.

```python
system = FirstOrderSystem(K=50, a=50)

print(
    f"time constant = {system.tau.to('ms'):~P.0f}",
    f"settling time = {system.settling_time.to('ms'):~P.3f}",
    f"rise time = {system.rise_time.to('ms'):~P.3f}",
    sep='\n'
)
```

```python
system.plot_unit_step_response(upper_limit=1)
```
